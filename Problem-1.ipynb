# 1D Random walk

We will concider two kinds of 1D random talk.
* $dx$ is uniformly distributed between $[-1/2, +1/2]$.
* $dx=+1$ with probability 1/2 and $dx=-1$ with probability 1/2.

For both kinds of the 1D random walk, you should do the following:
* Plot a single trajectory.
* Plot the histogram of the end point for an ensemble of trajectories.

## $dx$ is uniformly distributed between $[-1/2, +1/2]$

```python
import random 
import numpy as np 
import matplotlib.pyplot as plt 

step_n = 10000
path = np.random.uniform(-0.5, 0.5, step_n)
path = np.cumsum(path)
plt.title("single path of Random walk")
plt.xlabel('step')
plt.ylabel('location')
plt.plot(path)
```

```python
path_end = []
for i in range(6000):  #6000 times
    step_n = 10000  #each time 10000 steps
    path = np.random.uniform(-0.5, 0.5, step_n)
    path_end.append(np.sum(path))

plt.title("final location of all Random walk")
plt.xlabel('episode')
plt.ylabel('final location')
plt.hist(path_end)
```

## $dx=+1$ with probability 1/2 and $dx=-1$ with probability 1/2.

```python
dims = 1
step_n = 10000
step_set = [-1, 1]
origin = np.zeros((1,dims))
step_shape = (step_n,dims)
steps = np.random.choice(a=step_set, size=step_shape, p=[0.5, 0.5])

path = np.concatenate([origin, steps]).cumsum(0)
plt.title("single path of Random walk")
plt.xlabel('step')
plt.ylabel('location')
plt.plot(path)
```

```python
dims = 1
path_end = []
for i in range(6000):  #6000 times
    step_n = 10000  #each time 10000 steps
    step_set = [-1, 1]
    origin = np.zeros((1,dims))
    step_shape = (step_n,dims)
    steps = np.random.choice(a=step_set, size=step_shape, p=[0.5, 0.5])
    path = np.concatenate([origin, steps]).cumsum(0)
    path_end.append(path[-1][0])

plt.title("final location of all Random walk")
plt.xlabel('episode')
plt.ylabel('final location')
plt.hist(path_end)
```
