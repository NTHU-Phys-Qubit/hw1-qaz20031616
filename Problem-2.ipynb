# 2D Random walk

We will concider three kinds of 2D random talk.
* $dx$ and $dy$ are uniformly distributed between $[-1/2, +1/2]$.
* $dx=+1$ with probability 1/2 and $dx=-1$ with probability 1/2. Similarly for $dy$.
* $(dx, dy)=(\cos(\theta), \sin(\theta))$ where $\theta$ is uniformly distributed between $[0, 2\pi]$.

For all kinds of the 2D random walk, you should do the following:
* Plot a single trajectory.
* Plot the distribution of the end point for an ensemble of trajectories.
* Plot the histogram of the $r=x^2_f+y^2_f$ for an ensemble of trajectories.

## $dx$ and $dy$ are uniformly distributed between $[-1/2, +1/2]$.

```python
import random 
import numpy as np 
import matplotlib.pyplot as plt 
import math
step_n = 10000
path_x = np.random.uniform(-0.5, 0.5, step_n)
path_x = np.cumsum(path_x)
path_y = np.random.uniform(-0.5, 0.5, step_n)
path_y = np.cumsum(path_y)

plt.plot(path_x, path_y)
```

```python
path_end_x = []
path_end_y = []
for i in range(6000):  #6000 times
    step_n = 10000  #each time 10000 steps
    path_x = np.random.uniform(-0.5, 0.5, step_n)
    path_end_x.append(np.sum(path_x))
    path_y = np.random.uniform(-0.5, 0.5, step_n)
    path_end_y.append(np.sum(path_y))
    
plt.plot(path_end_x, path_end_y,'o')
```

```python
path_end = []

for i in range(6000):  #6000 times
    step_n = 10000  #each time 10000 steps
    path_x = np.random.uniform(-0.5, 0.5, step_n)
    x = np.sum(path_x)
    path_y = np.random.uniform(-0.5, 0.5, step_n)
    y = np.sum(path_y)
    path_end.append(math.sqrt((x**2)+(y**2)))
    
plt.hist(path_end)
```

## $dx=+1$ with probability 1/2 and $dx=-1$ with probability 1/2. Similarly for $dy$.

```python
dims = 2
step_n = 10000
step_set = [-1, 1]
origin = np.zeros((1,dims))
step_shape = (step_n,dims)
steps = np.random.choice(a=step_set, size=step_shape, p=[0.5, 0.5])
path = np.concatenate([origin, steps]).cumsum(0)

plt.plot(path[:,0],path[:,1])
```

```python
dims = 2
path_end_x = []
path_end_y = []
for i in range(6000):  #6000 times
    step_n = 10000  #each time 10000 steps
    step_set = [-1, 1]
    origin = np.zeros((1,dims))
    step_shape = (step_n,dims)
    steps = np.random.choice(a=step_set, size=step_shape, p=[0.5, 0.5])
    path = np.concatenate([origin, steps]).cumsum(0)
    path_end_x.append(path[-1][0])
    path_end_y.append(path[-1][1])
    
plt.plot(path_end_x, path_end_y,'o')
```

```python
dims = 2
path_end = []
for i in range(6000):  #6000 times
    step_n = 10000  #each time 10000 steps
    step_set = [-1, 1]
    origin = np.zeros((1,dims))
    step_shape = (step_n,dims)
    steps = np.random.choice(a=step_set, size=step_shape, p=[0.5, 0.5])
    path = np.concatenate([origin, steps]).cumsum(0)
    path_end.append(math.sqrt((path[-1][0]**2)+(path[-1][1]**2)))
    
plt.hist(path_end)
```

## $(dx, dy)=(\cos(\theta), \sin(\theta))$ where $\theta$ is uniformly distributed between $[0, 2\pi]$.

```python
step_n = 10000

path_x = np.cos(np.radians(np.random.uniform(0, 360, step_n)))
path_y = np.sin(np.radians(np.random.uniform(0, 360, step_n)))

path_x = np.cumsum(path_x)
path_y = np.cumsum(path_y)

plt.plot(path_x, path_y)
```

```python
path_end_x = []
path_end_y = []
for i in range(6000):  #6000 times
    step_n = 10000  #each time 10000 steps
    path_x = np.cos(np.radians(np.random.uniform(0, 360, step_n)))
    path_end_x.append(np.sum(path_x))
    path_y = np.sin(np.radians(np.random.uniform(0, 360, step_n)))
    path_end_y.append(np.sum(path_y))
    
plt.plot(path_end_x, path_end_y,'o')
```

```python
path_end = []

for i in range(6000):  #6000 times
    step_n = 10000  #each time 10000 steps
    path_x = np.cos(np.radians(np.random.uniform(0, 360, step_n)))
    x = np.sum(path_x)
    path_y = np.sin(np.radians(np.random.uniform(0, 360, step_n)))
    y = np.sum(path_y)
    path_end.append(math.sqrt((x**2)+(y**2)))
    
plt.hist(path_end)
```
